```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 14
plt.rcParams['font.family'] = 'serif'
```

```python
prefix = './dados/dados experimentais/Data2025mstar26number'
prefixU = './dados/dados experimentais/Ur2025mstar26number'

np.random.seed(42)
all_idxs = list(5*np.random.permutation(list(range(38))) + 95)
all_idxs.remove(190)
Y = np.zeros((len(all_idxs), 18000))
T = np.zeros((len(all_idxs), 18000))
U = np.zeros(len(all_idxs))
all_n = len(all_idxs)
train_idxs = all_idxs[:25]
test_idxs = all_idxs[25:]

for i, idx in enumerate(all_idxs):
    if idx == 190:
        pass
    else:
      dados = np.loadtxt(prefix + str(idx), delimiter=',')
      dadosU = np.loadtxt(prefixU + str(idx))
      Y[i, :] = dados[:, 1]
      T[i, :] = dados[:, 0]
      U[i] = np.mean(dadosU)
```

```python
rms_experimental = np.array([np.sqrt(np.mean(Y[i, :]**2)) for i in range(all_n)])

plt.figure(figsize=(12, 6))
plt.scatter(U, rms_experimental, color='blue')
plt.ylim(0, .7)
plt.xlim(3.5, 11)
plt.xlabel('Reduced velocity ($U_r$)')
plt.ylabel('RMS ( ${\overline{y^2}}$ )')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
